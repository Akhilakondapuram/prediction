```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("vgsales.csv")  # Replace with your actual file path if needed

# Drop rows with missing values in required columns
df = df.dropna(subset=["Global_Sales", "NA_Sales"])

# Define feature and target
X = df[['Global_Sales']]
y = df['NA_Sales']

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.9, random_state=42)

# Train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict and evaluate
y_pred = model.predict(X_test)
rmse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print model equation and performance
print(f"Root Mean Squared Error (RMSE): {rmse:.3f}")
print(f"RÂ² Score: {r2:.3f}")


# Optional: Visualization
plt.scatter(X_test, y_test, color='black', label='Actual NA Sales')
plt.plot(X_test, y_pred, color='red', label='Predicted Regression Line')
plt.xlabel('Global Sales')
plt.ylabel('NA Sales')
plt.title('NA Sales vs Global Sales')
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('vgsales.csv')

# Drop missing values
df = df.dropna(subset=["Global_Sales", "EU_Sales"])

# Define feature and target
X = df[['Global_Sales']]
y = df['EU_Sales']

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.9, random_state=42)

# Train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# Evaluate the model
print(f"RMSE: {mean_squared_error(y_test, y_pred):.4f}")
print(f"RÂ² Score: {r2_score(y_test, y_pred):.4f}")

# Plotting
plt.figure(figsize=(8, 6))
plt.scatter(X_test, y_test, color='blue', label='Actual EU Sales')
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Predicted Regression Line')
plt.xlabel('Global Sales (in millions)')
plt.ylabel('EU Sales (in millions)')
plt.title('EU Sales vs Global Sales')
plt.legend()
plt.grid(True)
plt.show()
```
